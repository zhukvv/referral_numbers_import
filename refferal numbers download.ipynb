1. Обязательно активировать в первую очередь:

```python
def func_data_export(dataFrame_loaded):

  #переименование стобцов
  dataFrame_loaded.columns=list(range(len(list(dataFrame_loaded.columns))))

  #удаление пустых стобцов
  for col in range(len(list(dataFrame_loaded.columns))):
    if dataFrame_loaded[col].isna().sum()==dataFrame_loaded.shape[0]:
      dataFrame_loaded.drop([col],axis=1, inplace=True)

  #восстановление очередности столбцов(их наименований)
  dataFrame_loaded.columns=list(range(len(list(dataFrame_loaded.columns))))

  #заполнение пропусков
  dataFrame_loaded.fillna("0", inplace=True)

  #удаление ненужных строк из УПД
  index = dataFrame_loaded[dataFrame_loaded[0].str.contains('А')].index[0]
  dataFrame_loaded = dataFrame_loaded.iloc[index:]
  dataFrame_loaded.reset_index(drop=True,inplace=True)

  #сделать первую строку наименованием колонок
  dataFrame_loaded.columns=dataFrame_loaded.iloc[0]
  dataFrame_loaded= dataFrame_loaded[1:]
  dataFrame_loaded.reset_index(drop=True, inplace=True)

  #оставляем только нужные стобцы, всегда именно такое наименование
  dataFrame_loaded = dataFrame_loaded[['А','1','3','4','5']]

  #удаляем все строки с "А"
  h=dataFrame_loaded['А'].apply(lambda x: x.find('А'))
  h= h[h==0]
  to_drop = h.index
  to_drop_list= list(to_drop)
  dataFrame_loaded.drop(index=to_drop_list, inplace=True)
  dataFrame_loaded.reset_index(drop=True, inplace=True)

  #удаляем все строки, которые не являются инфо об артикулах

  for num in range(len(dataFrame_loaded)):
    try:
      dataFrame_loaded['1'][num] = int(dataFrame_loaded['1'][num])
    except:
      dataFrame_loaded.drop(index=[num], inplace = True)

  dataFrame_loaded.reset_index(drop=True, inplace=True)

  for num in range(len(dataFrame_loaded)):
    if dataFrame_loaded['1'][num] == 0:
      dataFrame_loaded.drop(index=[num],inplace = True)

  dataFrame_loaded.reset_index(drop=True, inplace=True)

  # переименование столбцов
  dict_to_rename_columns = {'А':'ref','1':'#', '3':'qty', '4':'price_pcs','5':'price_total_woVAT'}
  dataFrame_loaded.rename(columns=dict_to_rename_columns, inplace=True)

  # назначение типов данных столбцов
  dataFrame_loaded['#']=dataFrame_loaded['#'].astype('int')
  dataFrame_loaded['qty']=dataFrame_loaded['qty'].astype('int')
  dataFrame_loaded['price_pcs']=dataFrame_loaded['price_pcs'].astype('float')
  dataFrame_loaded['price_total_woVAT']=dataFrame_loaded['price_total_woVAT'].astype('float')

  return dataFrame_loaded
```
